```python
import numpy as np, pandas as pd, os, sqlite3, json
```

```python
def read_json(file_path):
    obj_list = []
    with open(file_path, 'r') as file:
        for line in file:
                json_obj = json.loads(line)
                obj_list.append(json_obj)
    return obj_list;
```

```python
# Function to create SQLite database and tables
def create_database_and_tables():
    # Connect to SQLite database (or create if it doesn't exist)
    conn = sqlite3.connect('datalake.db')
    cursor = conn.cursor()

    # Create tables
    cursor.execute('''
    CREATE TABLE IF NOT EXISTS users (
        _id STRING PRIMARY KEY,
        active BOOLEAN,
        createdDate DATE,
        lastLogin DATE,
        role TEXT,
        signUpSource TEXT,
        state TEXT
    )
    ''')

    cursor.execute('''
    CREATE TABLE IF NOT EXISTS receipts (
        id STRING PRIMARY KEY,
        bonusPointsEarned INTEGER,
        bonusPointsEarnedReason TEXT,
        createDate DATE,
        dateScanned DATE,
        modifyDate DATE,           
        pointsAwardedDate DATE,
        pointsEarned TEXT,
        purchaseDate DATE,
        purchasedItemCount INTEGER,
        rewardsReceiptItemList ARRAY, 
        rewardsReceiptStatus TEXT,
        totalSpent REAL,
        userId TEXT 
    )
    ''')


    cursor.execute('''
    CREATE TABLE IF NOT EXISTS brands (
        _id STRING PRIMARY KEY,
        barcode TEXT,
        brandCode TEXT,
        category TEXT,
        categoryCode TEXT,
        cpg TEXT,                       
        topBrand BOOLEAN,
        name TEXT
    )
    ''')

    conn.commit()
    conn.close()

```
